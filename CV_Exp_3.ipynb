```python
from google.colab import files
uploaded = files.upload()

import cv2
from google.colab.patches import cv2_imshow

# get filename safely
img_path = next(iter(uploaded))

img = cv2.imread(img_path)

if img is None:
    print("Error: Image not loaded. Check file name/format.")
else:
    cv2_imshow(img)
```

```python
resized = cv2.resize(img, None, fx=0.5, fy=0.5)  # half size
cv2_imshow(resized)
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 degrees
rotated = cv2.warpAffine(img, matrix, (w, h))

cv2_imshow(rotated)
```

```python
crop = img[50:200, 50:200]
cv2_imshow(crop)
```

```python
horizontal = cv2.flip(img, 1)
vertical = cv2.flip(img, 0)

cv2_imshow(horizontal)
cv2_imshow(vertical)
```

```python
# alpha = contrast, beta = brightness
adjusted = cv2.convertScaleAbs(img, alpha=1.5, beta=50)

cv2_imshow(adjusted)
```
